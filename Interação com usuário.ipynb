Type

```python
x = 10
type(x)
print('x,', type(x))

nome = 'João Marques'
print('nome', type(nome))

num = '100'
num_conver = int(num)    # Converter Variável
print('num', type(num))
print('num_conver', type(num_conver))

num_float = float(num)
print('num_float', num_float)
```
