> Aluno Aprovados

```python
# Notas Aluno 1

pb_aluno1 = 8
sb_aluno1 = 7
tb_aluno1 = 6
qb_aluno1 = 10

soma = pb_aluno1 + sb_aluno1 + tb_aluno1 + qb_aluno1

print('Total:', (soma))

# Cálculo da Média

print('Média Final:', soma / 4)

# Verificação de Aprovado / Reprovado

if soma / 4 >= 7:
  print('Aprovado')

else:
  print('Reprovado')
```

# Diferença de Notas

> Utilizaremos operador *Subtração* para verificar a diferença de entre as notas

> Utilizaremos operador *Multiplicação* para multiplcar as notas finais

```python
# Diferença Aluno 1 e Aluno 2

aluno1 = 31

aluno2 = 29

diferenca = aluno1 - aluno2

print('Diferença de notas:', diferenca)

multiplicacao = diferenca * 4

print('Multiplicação:', multiplicacao)
```

```python
# Diferença Aluno 3 e Aluno 5

aluno3 = 25

aluno5 = 32

diferenca = aluno3 - aluno5

print('Diferença de notas:', diferenca)

multiplicacao = diferenca * 4

print('Multiplicação:', multiplicacao)
#
```
